```python
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#Read Data

data=pd.read_csv('/content/USA_Housing.csv')
data
```

```python
#split column

columns=['Avg. Area Number of Rooms','Price']
columns
```

```python
new_data=data[columns]
new_data
```

```python
#labeling data
x=new_data['Avg. Area Number of Rooms']
y=new_data['Price']
```

```python
#Split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/4,random_state=0,shuffle=True)
```

```python
x_train=np.array(x_train).reshape(-1,1)
```

```python
x_test=np.array(x_test).reshape(-1,1)
x_test
```

```python
y_train=np.array(y_train).reshape(-1,1)
```

```python
y_test=np.array(y_test).reshape(-1,1)
```

```python
#Train model
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
#Prediction
y_pred=model.predict(x_test)
y_pred
```

```python
#Plotting
plt.scatter(x_test,y_test,color='Yellow')
plt.plot(x_test,model.predict(x_test),color='Red')
plt.title('USA House Price Prediction Model')
plt.xlabel('Avg. Area Number of Rooms')
plt.ylabel('Prices($millions)')
plt.show()
```
